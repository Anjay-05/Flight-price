```python
#importing libraries
import pandas as pd
import time
from selenium import webdriver
from selenium.common.exceptions import NoSuchElementException as e
from selenium.common.exceptions import ElementClickInterceptedException as c
import requests
```

```python
#Connecting to web driver
driver=webdriver.Chrome(r"C:\Users\user\Downloads\chromedriver_win32\chromedriver.exe")

#Getting the web page
url="https://www.yatra.com/"
driver.get(url)
```

```python
search=driver.find_element_by_xpath('//div[@class="ripple-parent search-height demo-icon icon-go"]/input')
search.click()
```

```python
Airline=[]
Date=[]
Source=[]
Destination=[]
Duration=[]
Dep_time=[]
Arr_time=[]
Totalstops=[]
Price=[]
Num=[]
```

```python
def flightA():
    #Scraping Airline names
    try:
        airline=driver.find_elements_by_xpath('//div[@class="fs-13 airline-name no-pad col-8"]/span')
        for i in airline:
            Airline.append(i.text)
    except e:
        Airline.append('-')

    #Departure time
    try:
        deptime=driver.find_elements_by_xpath('//div[@class="i-b pr"]')
        for i in deptime:
            Dep_time.append(i.text)
    except e:
        Dep_time.append('-')
        
    #Arrival time
    try:
        arrtime=driver.find_elements_by_xpath('//p[@class="bold fs-15 mb-2 pr time"]')
        for i in arrtime:
            Arr_time.append(i.text.split('\n')[0])
    except e:
        Arr_time.append('-')
    
    #Source
    try:
        source=driver.find_elements_by_xpath('//div[@class="i-b col-4 no-wrap text-right dtime col-3"]')
        for i in source:
            Source.append(i.text.split('\n')[1])
    except e:
        Source.append('-')
        
    #Destination
    try:    
        desti=driver.find_elements_by_xpath('//div[@class="i-b pdd-0 text-left atime col-5"]')
        for i in desti:
            Destination.append(i.text.split('\n')[-1])
    except e:
        Destination.append('-')
        
    #Duration
    try:
        dur=driver.find_elements_by_xpath('//div[@class="stop-cont pl-13"]/p')
        for i in dur:
            Duration.append(i.text)
    except e:
        Duration.append('-')

    #Total stops
    try:
        stop=driver.find_elements_by_xpath('//div[@class=" font-lightgrey fs-10 tipsy i-b fs-10"]/span')
        for i in stop:
            Totalstops.append(i.text)
    except e:
        Totalstops.append('-')

    #Price
    try:
        price=driver.find_elements_by_xpath('//div[@class="i-b tipsy fare-summary-tooltip fs-18"]')
        for i in price:
            Price.append(i.text)
    except e:
        Price.append('-')

    #Date
    try:
        da=driver.find_element_by_xpath('//div[@class="link-color fs-12 cursor-pointer"]')
        da.click()

        Num.clear()
        num=driver.find_elements_by_xpath('//div[@class="fs-13 airline-name no-pad col-8"]/p')
        for i in num:
            Num.append(i.text)
        date=driver.find_elements_by_xpath('//div[@class="tab-content scroll-v"]/div[1]/div[2]/div[1]/div[3]')
        for i in date:
            for j in range(len(Num)):
                Date.append(i.text.split(',')[1])
    except e:
        Date.append('-')
```

```python
flightA()
```

```python
len(Airline),len(Date),len(Source),len(Destination),len(Dep_time),len(Arr_time),len(Duration),len(Totalstops),len(Price)
```

```python
df=pd.DataFrame({'Airline':Airline,'Date of Journey':Date,'Source':Source,'Destination':Destination,'Departure Time':Dep_time,
                 'Arrival Time':Arr_time,'Duration':Duration,'Total Stops':Totalstops,'Price':Price})
df
```

```python
#Connecting to web driver
driver=webdriver.Chrome(r"C:\Users\user\Downloads\chromedriver_win32\chromedriver.exe")

#Getting the web page
url="https://www.makemytrip.com/"
driver.get(url)
```

```python
ad=driver.find_element_by_xpath('//li[@class="makeFlex hrtlCenter font10 makeRelative lhUser userLoggedOut"]')
ad.click()
time.sleep(3)
#Search button clicking
search_btn=driver.find_element_by_xpath('//a[@class="primaryBtn font24 latoBold widgetSearchBtn "]')
search_btn.click()
```

```python
_Airline=[]
_Date=[]
_Source=[]
_Destination=[]
_Duration=[]
_Dep_time=[]
_Arr_time=[]
_Totalstops=[]
_Price=[]
```

```python
def flightB():
    #Selecting view all prices
    view=driver.find_element_by_xpath('//div[@class="makeFlex top gap-x-10"]/button')
    view.click()

    #Scraping Airline names
    try:
        _airline=driver.find_elements_by_xpath('//span[@class="boldFont blackText airlineName"]')
        for i in _airline:
            _Airline.append(i.text)
    except e:
        _Airline.append('-')
    #Departure time
    DT=[]
    _deptime=driver.find_elements_by_xpath('//p[@class="blackText fontSize18 blackFont appendBottom2 makeFlex hrtlCenter"]')
    for i in _deptime:
        DT.append(i.text)

    for i in range(0,len(DT),2):
        _Dep_time.append(DT[i])
    #Arrival time    
    for i in range(1,len(DT),2):
        _Arr_time.append(DT[i])

    #Source
    SO=[]
    _source=driver.find_elements_by_xpath('//p[@class="darkText"]')
    for i in _source:
        SO.append(i.text)

    for i in range(0,len(SO),2):
            _Source.append(SO[i])
    #Destination
    for i in range(1,len(SO),2):
            _Destination.append(SO[i])

    #Duration
    try:

        _dur=driver.find_elements_by_xpath('//div[@class="stop-info flexOne"]/p')
        for i in _dur:
            _Duration.append(i.text)
    except e:
        _Duration.append('-')
        
    #Total stops
    try:
        _stop=driver.find_elements_by_xpath('//div[@class="stop-info flexOne"]/div')
        for i in _stop:
            _Totalstops.append(i.text.split('via')[0])
    except e:
        _Totalstops.append('-')

    #Price
    try:
        _price=driver.find_elements_by_xpath('//div[@class="textRight flexOne"]')
        for i in _price:
            _Price.append(i.text)
    except e:
        _Price.append('-')

    #selecting flight details
    flt=driver.find_element_by_xpath('//span[@class="linkText ctaLink"]')
    flt.click()
    time.sleep(3)
    #Date
    day=[]
    date=driver.find_elements_by_xpath('//div[@class="airlineDTInfoCol"]/p[2]')
    for i in date:
        day.append(i.text.split(','))
    day=day[1][1]

    Ran=[]
    try:
        ran=driver.find_elements_by_xpath('//span[@class="arln-logo logo1"]')
        for i in ran:
            Ran.append(i.text)
    except e:
        Ran.append('-')
    for i in range(len(Ran)):
        _Date.append(day)
```
